```python
# headers = 0 saying, first row is the header

file1 = '/content/drive/MyDrive/MY Startups/Misc/Domain Name valuation/Data/Cleaned Data/Sales_Prices_Category.csv'
file2 = '/content/drive/MyDrive/MY Startups/Misc/Domain Name valuation/Data/Cleaned Data/category1.txt'
file3 = '/content/drive/MyDrive/MY Startups/Misc/Domain Name valuation/Data/Cleaned Data/more_sales_data.csv'
myPath = '/content/drive/MyDrive/MY Startups/Misc/Domain Name valuation/Data/Cleaned Data/'

dataset1 = pd.read_csv(file1, header=0)
dataset2 = pd.read_csv(file2, header=0)
dataset3 = pd.read_csv(file3, header=0)

dataset1.head

dataset1.to_csv(myPath + 'cat2to10_new.csv')
```

```python
# Print some stats
count_row_data1 = dataset1.shape[0] 
count_row_data2 = new_dataset2.shape[0] 
count_row_data3 = dataset3.shape[0] 

total_before_merge = count_row_data1 + count_row_data2 + count_row_data3


print('--------------------------Daataset1----------------------')
print(dataset1.head)
print('--------------------------Daataset2----------------------')
print(new_dataset2.head)
print('--------------------------Daataset3----------------------')
print(dataset3.head)


result_inter = dataset1.append(new_dataset2, ignore_index=True)
print('--------------------------Daataset inter ----------------------')
print(result_inter.head)

result_final = result_inter.append(dataset3, ignore_index=True)
print('--------------------------Daataset final ----------------------')
print(result_final.head)



# save to csv
result_final.to_csv(myPath + 'DomainMergedFile.csv')

count_row_data_inter_Merge = result_final.shape[0] 

print('count_row_data1  =', count_row_data1  )
print('count_row_data2  =', count_row_data2  )
print('count_row_data3  =', count_row_data3  )
print('count_row_data_All  =', count_row_data3 +count_row_data2 + count_row_data1  )
print('count_row_data_Merged  =', count_row_data_inter_Merge  )
```
